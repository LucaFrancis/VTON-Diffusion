```python
import matplotlib.pyplot as plt
import pandas as pd
embedding_name = "christphgrmmr"
embedding_name_abbrv = "Embedding D"
df = pd.read_csv("embedding_datasets/"+embedding_name+"/results/face_study.csv", sep=',', header=None)
    
df[0] = df[0].transform(lambda path: int(path.split("step")[1].split(".")[0]), axis=0)
df = df.sort_values(by=[0])
df = df.groupby(0).mean().reset_index()
df
```

```python
inverted_df = df
inverted_df[1] = inverted_df[1].transform(lambda face_distance: 1-face_distance, axis=0)
```

```python
plt.xlim(0, 3000)
plt.ylim(0, 1)
face_distance_threshold = 0.4
face_similarity_threshold = 1 - face_distance_threshold
plt.axhline(face_similarity_threshold, linestyle="--", color="r", label="Face detection threshold")
plt.legend()
plt.title(embedding_name_abbrv+" - face similarity during training")
plt.xlabel("Textual inversion training step")
plt.plot(inverted_df[0], inverted_df[1],)
plt.savefig("./plots/face_study_"+embedding_name+".png")
```
