```python
from mpl_toolkits.axes_grid1 import ImageGrid
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(4., 10.))
grid = ImageGrid(fig, 111, nrows_ncols=(3, 4), axes_pad=0.01, )

continuity1 =  plt.imread('artifact_examples/Continuity/05_1_front.jpg_00001_.png')
continuity2 =  plt.imread('artifact_examples/Continuity/08_1_front.jpg_00001_.png')
noisy_generation1 =  plt.imread('artifact_examples/noisy_generation/05_1_front.jpg_00001_.png')
noisy_generation2 =  plt.imread('artifact_examples/noisy_generation/05_1_front.jpg_00001_2.png')
noisy_generation3 =  plt.imread('artifact_examples/noisy_generation/06_2_side.jpg_00001_.png')
noisy_generation4 =  plt.imread('artifact_examples/noisy_generation/34_1_front.jpg_00001_.png')
side_view1 =  plt.imread('artifact_examples/side_view/03_2_side.jpg_00001_.png')
side_view2 =  plt.imread('artifact_examples/side_view/14_2_side.jpg_00001_.png')
side_view3 =  plt.imread('artifact_examples/side_view/16_2_side.jpg_00001_.png')
side_view4 =  plt.imread('artifact_examples/side_view/17_2_side.jpg_00001_.png')
genderswap1 =  plt.imread('artifact_examples/genderswap/06_1_front.jpg_00001_.png')
genderswap2 =  plt.imread('artifact_examples/genderswap/06_1_front.jpg_00001_2.png')

for ax, im in zip(grid,
                  [ continuity1, continuity2,genderswap1,genderswap2,
                   noisy_generation1, noisy_generation2, noisy_generation3, noisy_generation4,
                   side_view1, side_view2, side_view3, side_view4,
                   ]):
    if im is not None:
        ax.imshow(im)
    ax.axis("off")
# Display the figure
plt.savefig('plots/artifacts.png', format='png', dpi=1000, bbox_inches='tight')
plt.show()
```
