```python
from mpl_toolkits.axes_grid1 import ImageGrid
import matplotlib.pyplot as plt

dataset_path = "datasets/deep-fashion/In-shop Clothes Retrieval Benchmark/Img/"

path1 = 'img_highres/MEN/Jackets_Vests/id_00000084/08_1_front.jpg'
path2 = 'img_highres/MEN/Sweaters/id_00001919/03_1_front.jpg'
path3 = 'img_highres/MEN/Tees_Tanks/id_00000755/05_1_front.jpg'
path4 = 'img_highres/MEN/Denim/id_00003074/01_1_front.jpg'

path5 = "img_highres/WOMEN/Tees_Tanks/id_00006777/01_1_front.jpg"
path6 = "img_highres/WOMEN/Tees_Tanks/id_00007708/03_1_front.jpg"
path7 = "img_highres/WOMEN/Blouses_Shirts/id_00007072/02_1_front.jpg"
path8 = "img_highres/WOMEN/Sweatshirts_Hoodies/id_00007908/01_1_front.jpg"
# Load the images
embedding_A = plt.imread('embedding_datasets/lucfrncs/img/processed/00016-0-6.png')
embedding_B = plt.imread('embedding_datasets/frankfrschlng/img/processed/00003-0-10.png')
embedding_C = plt.imread('embedding_datasets/fabhsr/img/processed/00008-0-IMG_4117.png')
embedding_D = plt.imread('embedding_datasets/christphgrmmr/img/processed/00000-0-PXL_20230920_124959864.png')


def get_transformed_images(embedding_name: str):
    transform_all_path = "embedding_datasets/" + embedding_name + "/results/transform-all/"
    img1 = plt.imread(transform_all_path + path1 + "_00001_.png")
    img2 = plt.imread(transform_all_path + path2 + "_00001_.png")
    img3 = plt.imread(transform_all_path + path3 + "_00001_.png")
    img4 = plt.imread(transform_all_path + path4 + "_00001_.png")
    img5 = plt.imread(transform_all_path + path5 + "_00001_.png")
    img6 = plt.imread(transform_all_path + path6 + "_00001_.png")
    img7 = plt.imread(transform_all_path + path7 + "_00001_.png")
    img8 = plt.imread(transform_all_path + path8 + "_00001_.png")
    return img1, img2, img3, img4, img5, img6, img7, img8


test_1 = plt.imread(
    dataset_path + path1)
test_2 = plt.imread(
    dataset_path + path2)
test_3 = plt.imread(
    dataset_path + path3)
test_4 = plt.imread(
    dataset_path + path4)

test_5 = plt.imread(
    dataset_path + path5)
test_6 = plt.imread(
    dataset_path + path6)
test_7 = plt.imread(
    dataset_path + path7)
test_8 = plt.imread(
    dataset_path + path8)

A1, A2, A3, A4, A5, A6, A7, A8 = get_transformed_images("lucfrncs")
B1, B2, B3, B4, B5, B6, B7, B8 = get_transformed_images("frankfrschlng")
C1, C2, C3, C4, C5, C6, C7, C8 = get_transformed_images("fabhsr")
D1, D2, D3, D4, D5, D6, D7, D8 = get_transformed_images("christphgrmmr")
```

```python
fig = plt.figure(figsize=(5., 5.))
grid = ImageGrid(fig, 111, nrows_ncols=(5, 5), axes_pad=0.1, )

for ax, im in zip(grid,
                  [None, embedding_A, embedding_B, embedding_C, embedding_D,
                   test_1, A1, B1, C1, D1,
                   test_2, A2, B2, C2, D2,
                   test_3, A3, B3, C3, D3,
                   test_4, A4, B4, C4, D4,
                   ]):
    # Iterating over the grid returns the Axes.
    if im is not None:
        ax.imshow(im)
    ax.axis("off")
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height
grid[0].text(1, 0.5 * (bottom + top), 'embedding',
             horizontalalignment='center',
             verticalalignment='center',
             rotation='vertical',
             transform=grid[0].transAxes,
             size=5.)
grid[0].text(0.5 * (left + right), 0, 'test image',
             horizontalalignment='center',
             verticalalignment='center',
             transform=grid[0].transAxes,
             size=5.)
# Display the figure
plt.savefig('plots/transformed_examples_MAN.png', format='png', dpi=1000, bbox_inches='tight')
plt.show()
```
